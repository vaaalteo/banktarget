```python
#importing libaries to use
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read BankChurners file
bankdata = pd.read_csv('BankChurners.csv')

#load the bankdata into dataframe
bankdf = pd.DataFrame(bankdata) 
bankdf = bankdf.set_index("CLIENTNUM") #setting the first column as index
#print(bankdf)

print("This is the shape of the dataset")
print(bankdf.shape)
print()
print("This is the index of the dataset")
print(bankdf.index)
print()
print("These are the columns in the dataset")
print(bankdf.columns)
print()
print("The total number of non-NA values in this dataset is:")
print(bankdf.count())
print()
print("A summary of this dataset is shown below:")
print(bankdf.info())
print()
print("A descriptive statistical summary of this dataset is shown below:")
#print(bankdf['Total_Trans_Amt'].describe())
print(bankdf.describe())
```

```python
#dropping of irrelevant columns from the data
bankdf_revised = bankdf.drop(['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'],axis=1)
```

```python
#setting of the figure's dimension
fig_dims = (10, 7)

#plotting boxplot with seaborn and matplotlib
fig, ax = plt.subplots(figsize=fig_dims)
sns.boxplot(x=bankdf_revised['Marital_Status'], y=bankdf_revised["Total_Trans_Amt"],ax=ax, palette= 'pastel', width = 0.4, fliersize = 3)
plt.title('Credit Card Spending by Marital Status')
plt.show()

```
